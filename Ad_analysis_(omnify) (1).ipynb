```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel("/content/Omnify-Ad-data (G.A and L.S).xlsx")

df.head()
```

```python
#cleaning and converting data into appropriate format

df['Week'] = pd.to_datetime(df['Week'])
df['Payment Date'] = pd.to_datetime(df['Payment Date'])
df['Cost ($)'] = df['Cost ($)'].replace('[\$,]','',regex=True).astype(float)

df['Payment ($)'] = df['Payment ($)'].replace('[\$,]','',regex=True).astype(float)
df['Returns %'] = (df['Payment ($)']/df['Cost ($)'])

df.mean()
```

```python
#dealing with Null values
#df.fillna(mean, inplace=True)
df
#df.info()
```

### WEEKLY REPORT

```python
# Group the data by Week and Channel, and calculate key metrics
weekly_report = df.groupby([pd.Grouper(key='Week', freq='W')]) \
    .agg({
        'Cost ($)': 'sum',
        'Returns %': 'mean',
        'Impressions': 'sum',
        'Clicks': 'sum',
        'Prospects': 'sum',
        'Payment ($)': 'sum'
    }).reset_index()
```

### MONTHLY METRICS

```python
# Group the data by Week and Channel, and calculate key metrics
monthly_report = df.groupby([pd.Grouper(key='Week', freq='M')]) \
    .agg({
        'Cost ($)': 'sum',
        'Returns %': 'mean',
        'Impressions': 'sum',
        'Clicks': 'sum',
        'Prospects': 'sum',
        'Payment ($)': 'sum'
    }).reset_index()
```

```python
# Determine the most profitable category/keyword
most_profitable_keyword = df.groupby(['Search Keyword'])['Returns %'].mean().idxmax()
```

```python
weekly_report.to_csv("weekly_report.csv", index=False)
monthly_report.to_csv("monthly_report.csv", index=False)
```

```python
# Print the results
print("Weekly Report:")
print(weekly_report)

print("\nMonthly Report:")
print(monthly_report)


print("\nMost Profitable Keyword:")
print(most_profitable_keyword)
```

```python
df1 = pd.read_excel('/content/Omnify-Ad-data (G.A and L.S).xlsx')
df1.head()
```

```python
df1['Leads'] = df1['Leads'].fillna(0)
df1['Prospects'] = df1['Prospects'].fillna(0)
df1['Paid'] = df1['Paid'].fillna(0)
```

```python
df1['Date of Report'] = pd.to_datetime(df1['Date of Report'])
df1['Paid Date'] = pd.to_datetime(df1['Paid Date'])

df1['Money Spent ($)'] = df1['Money Spent ($)'].astype(float)
df1['Paid'] = df1['Paid'].replace('[\$,]','',regex=True).astype(float)

df1['Returns %'] = (df1['Paid']/df1['Money Spent ($)'])

df1.head()
```

```python
most_profitable_category_keyword = df1.groupby(['Categories'])['Returns %'].mean().idxmax()
most_profitable_category_keyword
```

```python
most_profitable_category_keyword = df1.groupby(['Categories'])['Prospects'].mean().idxmax()
most_profitable_category_keyword
```

```python
df1.head()
# Grouping the data by Week and Channel, and calculating key metrics
weekly_report1 = df1.groupby([pd.Grouper(key='Date of Report', freq='W'), 'Channel']) \
    .agg({
        'Money Spent ($)': 'sum',
        'Returns %': 'mean',
        'Clicks': 'sum',
        'Prospects': 'sum',
        'Paid': 'sum'
    }).reset_index()

print("Weekly Report:")
print(weekly_report1)
```

```python
# Grouping the data by Month and Channel, and calculating key metrics

monthly_report1 = df1.groupby([pd.Grouper(key='Date of Report', freq='M'), 'Channel']) \
    .agg({
        'Money Spent ($)': 'sum',
        'Returns %': 'mean',
        'Clicks': 'sum',
        'Prospects': 'sum',
        'Paid': 'sum'
    }).reset_index()

print("\nMonthly Report:")
print(monthly_report1)
```

```python
#the most profitable channel

channel_report = df1.groupby('Channel').agg({
     'Money Spent ($)': 'sum',
        'Returns %': 'mean',
        'Clicks': 'sum',
        'Prospects': 'sum',
        'Paid': 'sum'
}).reset_index()

channel_report.to_csv("channel_report.csv", index=False)


print("\nChannel Report:")
print(channel_report)


#the most profitable category/keyword

most_profitable_keyword = df1.groupby('Categories')['Returns %'].idxmax()


print("\nMost Profitable Keyword:")
print(most_profitable_keyword)
```

```python
# Saving the reports to separate CSV files
weekly_report1.to_csv("weekly_report1.csv", index=False)
monthly_report1.to_csv("monthly_report1.csv", index=False)



print("\nMost profitable channel and its statistics: ")
print(channel_report.loc[channel_report['Returns %'].idxmax()])
```
